```python
data1= data.drop(["customer name","customer e-mail","gender","age"],axis=1)
```

```python
x=data1.drop("car purchase amount",axis=1)
y=data1["car purchase amount"]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data1
```

```python
sc=StandardScaler()
```

```python
req_cols=['country', 'annual Salary', 'credit card debt', 'net worth',
       'car purchase amount']
```

```python
data1.loc[:,req_cols] = sc.fit_transform(data1[req_cols])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=134)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape, y_train.shape
```

```python
model = Sequential()
model.add(Dense(32, activation = "relu",input_shape = (4,)))
model.add(Dense(16, activation = "relu"))
model.add(Dense(1, activation = None))  
model.summary()
```

```python
model.compile(optimizer="adam",metrics = ["mse"], loss = "mean_squared_error")
```

```python
model.fit(x_train, y_train , epochs = 200, batch_size= 10
          )
```

```python
y_pred =model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
model1 = Sequential()
model1.add(Dense(32, activation = "linear",input_shape = (4,)))
model1.add(Dense(16, activation = "relu"))
model1.add(Dense(16, activation = "relu"))
model1.add(Dense(16, activation = "relu"))
model1.add(Dense(16, activation = "relu"))
model1.add(Dense(1, activation = "linear"))  
model1.summary()
```

```python
model1.compile(optimizer="adam",metrics = ["mse"], loss = "mean_squared_error")
```

```python
model1.fit(x_train, y_train , epochs = 200, batch_size= 10)
```

```python
model1.evaluate(x_test, y_test)
```
