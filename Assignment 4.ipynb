```python
x = dt[dt.columns]
```

## HC Clustering without PCA

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
mergings = linkage(x, method = "complete")
```

```python
plt.figure(figsize=(20,4))
dendrogram(mergings)
plt.show()
```

```python
labels = fcluster(mergings, 90, criterion='distance')
```

```python
labels
```

```python
dt['hc_cluster_without_pca'] = labels
```

```python
dt
```

## HC Clustering with PCA

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components=50)
```

```python
df = dt
```

```python
pc_arr = pc.fit_transform(df)
```

```python
data_pca = pd.DataFrame(pc_arr, columns = ['PC'+str(i) for i in range(1,51)])
```

```python
data_pca
```

```python
sum(pc.explained_variance_ratio_ *100)
```

```python
y = data_pca[data_pca.columns]
```

```python
mergings1 = linkage(y, method = "complete")
```

```python
plt.figure(figsize=(20,4))
dendrogram(mergings1)
plt.show()
```

```python
labels1 = fcluster(mergings1, 90, criterion='distance')
```

```python
labels1
```

```python
dt['hc_cluster_with_pca'] = labels1
```
